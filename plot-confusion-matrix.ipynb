```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define the confusion matrix
confusion_matrix = np.array([[114, 3, 0, 0],
                             [0, 118, 0, 0],
                             [0, 0, 117, 14],
                             [0, 0, 15, 84]])

# Create a heatmap of the confusion matrix
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  # Adjust font size if needed
sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d', square=True,
            xticklabels=["Class 1", "Class 2", "Class 3", "Class 4"],
            yticklabels=["Class 1", "Class 2", "Class 3", "Class 4"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")

# Display the plot
plt.show()

# Save the plot as an image (e.g., PNG)
plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')




# Define the confusion matrix
confusion_matrix = np.array([[114, 3, 0, 0],
                             [0, 118, 0, 0],
                             [0, 0, 117, 14],
                             [0, 0, 15, 84]])

# Calculate accuracy per class
accuracies = []

for i in range(len(confusion_matrix)):
    true_positive = confusion_matrix[i, i]
    total_samples = np.sum(confusion_matrix[i, :])
    accuracy = true_positive / total_samples
    accuracies.append(accuracy)

# Calculate average accuracy
average_accuracy = np.mean(accuracies)

# Print accuracy per class
for i, accuracy in enumerate(accuracies):
    print(f'Class {i + 1} Accuracy: {accuracy:.4f}')

# Print average accuracy
print(f'Average Accuracy: {average_accuracy:.4f}')

```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score

# Define the confusion matrix
confusion_matrix = np.array([[114, 3, 0, 0],
                             [0, 118, 0, 0],
                             [0, 0, 117, 14],
                             [0, 0, 15, 84]])

# Calculate precision, recall, and F1-score per class
class_labels = [f'Class {i + 1}' for i in range(len(confusion_matrix))]

precisions = [precision_score(confusion_matrix[i, :], confusion_matrix[:, i], average=None)[i] for i in range(len(confusion_matrix))]
recalls = [recall_score(confusion_matrix[i, :], confusion_matrix[:, i], average=None)[i] for i in range(len(confusion_matrix))]
f1_scores = [f1_score(confusion_matrix[i, :], confusion_matrix[:, i], average=None)[i] for i in range(len(confusion_matrix))]

# Calculate average precision, recall, and F1-score
avg_precision = np.mean(precisions)
avg_recall = np.mean(recalls)
avg_f1_score = np.mean(f1_scores)

# Create a Pandas DataFrame to store the results
results_df = pd.DataFrame({'Class': class_labels, 'Precision': precisions, 'Recall': recalls, 'F1-Score': f1_scores})

# Add a row for average values
avg_row = pd.DataFrame({'Class': ['Average'], 'Precision': [avg_precision], 'Recall': [avg_recall], 'F1-Score': [avg_f1_score]})
results_df = pd.concat([results_df, avg_row], ignore_index=True)

# Print the results in a table format
print(results_df)
```
